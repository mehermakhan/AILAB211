🔹 5. Modify Attributes

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

s = Student("Ali", 80)

print("Original:", s.name, s.marks)

# Modify
s.name = "Ahmed"
s.marks = 90

print("Updated:", s.name, s.marks)
```

🔹 6. Add & Delete Attributes

```python
class Student:
    def __init__(self, sid, name):
        self.sid = sid
        self.name = name

s = Student(1, "Ali")

# Add attribute
s.student_class = "BSCS"

print(s.__dict__)

# Delete attribute
del s.name

print(s.__dict__)
```

🔹 7. Display All Attributes Function

```python
class Student:
    def __init__(self, sid, name, student_class):
        self.sid = sid
        self.name = name
        self.student_class = student_class

    def display(self):
        for key, value in self.__dict__.items():
            print(key, ":", value)

s = Student(1, "Ali", "BSCS")
s.display()
```

🔹 8. Two Objects

```python
class Student:
    def __init__(self, sid, name, student_class):
        self.sid = sid
        self.name = name
        self.student_class = student_class

    def display(self):
        print("ID:", self.sid)
        print("Name:", self.name)
        print("Class:", self.student_class)
        print("-----------")

student1 = Student(1, "Ali", "BSCS")
student2 = Student(2, "Sara", "BBA")

student1.display()
student2.display()
```

🔹 10. Reduce Memory Usage (__slots__)

```python
class Student:
    __slots__ = ['sid', 'name', 'student_class']

    def __init__(self, sid, name, student_class):
        self.sid = sid
        self.name = name
        self.student_class = student_class

s = Student(1, "Ali", "BSCS")

import sys
print("Memory with slots:", sys.getsizeof(s))
```

✅ Lab Task 2: OOP Area Comparison
🔹 1. Classes for Shapes

```python
class Trapezoid:
    def __init__(self, a, b, h):
        self.a = a
        self.b = b
        self.h = h

    def area(self):
        return 0.5 * (self.a + self.b) * self.h


class Parallelogram:
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return self.base * self.height
```

🔹 2–5. Comparison + Dictionary + Iteration

```python
areas = {}

# Multiple inputs
data = [
    (5, 7, 4),   # trapezoid
    (6, 8, 5),
]

for i, (a, b, h) in enumerate(data):
    t = Trapezoid(a, b, h)
    p = Parallelogram(a, h)

    areas[f"Trapezoid_{i}"] = t.area()
    areas[f"Parallelogram_{i}"] = p.area()

# Display areas
print("Areas:")
for shape, value in areas.items():
    print(shape, "=", value)

# Find maximum
max_shape = max(areas, key=areas.get)
print("\nMaximum Area:", max_shape, "=", areas[max_shape])
```
