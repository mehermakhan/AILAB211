```python
# Graph representation using dictionary
graph = {
    '5': ['3', '7'],
    '3': ['2', '4'],
    '7': ['8'],
    '2': [],
    '4': ['8'],
    '8': []
}

# List to keep track of visited nodes
visited = []

# Queue for BFS
queue = []

# BFS function
def bfs(visited, graph, node):
    visited.append(node)
    queue.append(node)

    while queue:
        m = queue.pop(0)   # Remove from front
        print(m, end=" ")

        for neighbour in graph[m]:
            if neighbour not in visited:
                visited.append(neighbour)
                queue.append(neighbour)

# Driver Code
print("Following is the Breadth-First Search:")
bfs(visited, graph, '5')
```

```python
# Function to print the parent of each node (DFS)
def printParents(node, adj, parent):
    if parent == 0:
        print(node, "-> Root")
    else:
        print(node, "->", parent)

    for cur in adj[node]:
        if cur != parent:
            printParents(cur, adj, node)


# Function to print the children of each node (BFS)
def printChildren(Root, adj):
    q = []
    q.append(Root)

    vis = [0] * len(adj)

    while len(q) > 0:
        node = q.pop(0)
        vis[node] = 1

        print(node, "-> ", end=" ")

        for cur in adj[node]:
            if vis[cur] == 0:
                print(cur, end=" ")
                q.append(cur)
        print()


# Function to print leaf nodes
def printLeafNodes(Root, adj):
    for i in range(1, len(adj)):
        if len(adj[i]) == 1 and i != Root:
            print(i, end=" ")
    print()


# Function to print degrees of nodes
def printDegrees(Root, adj):
    for i in range(1, len(adj)):
        print(i, ":", end=" ")

        if i == Root:
            print(len(adj[i]))        # Root
        else:
            print(len(adj[i]) - 1)    # Other nodes


# ---------------- DRIVER CODE ----------------

N = 7
Root = 1

# Create adjacency list
adj = []
for i in range(N + 1):
    adj.append([])

# Creating the tree (edges)
adj[1].append(2)
adj[2].append(1)

adj[1].append(3)
adj[3].append(1)

adj[1].append(4)
adj[4].append(1)

adj[2].append(5)
adj[5].append(2)

adj[2].append(6)
adj[6].append(2)

adj[4].append(7)
adj[7].append(4)


# Output
print("The parents of each node are:")
printParents(Root, adj, 0)

print("\nThe children of each node are:")
printChildren(Root, adj)

print("\nThe leaf nodes of the tree are:")
printLeafNodes(Root, adj)

print("\nThe degrees of each node are:")
printDegrees(Root, adj)
```

```python
from collections import deque

# -------- Weighted Graph --------
graph = {
    'A': [('B', 2), ('C', 4)],
    'B': [('D', 7), ('E', 3)],
    'C': [('F', 5)],
    'D': [('G', 1)],
    'E': [('G', 6)],
    'F': [('G', 2)],
    'G': []
}

# -------- BFS Function --------
def bfs_with_cost(graph, start, goal):
    queue = deque()
    queue.append((start, [start], 0))   # (node, path, total_cost)

    visited = []

    while queue:
        node, path, cost = queue.popleft()

        # If goal found
        if node == goal:
            return path, cost

        if node not in visited:
            visited.append(node)

            for neighbour, weight in graph[node]:
                queue.append((neighbour, path + [neighbour], cost + weight))

    return None, 0


# -------- DRIVER CODE --------
print("BFS on Weighted Graph\n")

# Change these values to test different cases
start = 'A'
goal = 'G'

path, cost = bfs_with_cost(graph, start, goal)

# -------- OUTPUT --------
if path:
    print("Start Node:", start)
    print("Goal Node:", goal)
    print("Path:", " -> ".join(path))
    print("Total Cost:", cost)
else:
    print("No path found")
```
